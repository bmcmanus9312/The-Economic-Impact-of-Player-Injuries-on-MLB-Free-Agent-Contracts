```python
import pandas as pd

df = pd.read_csv(r"C:\Users\brend\Desktop\SABR_research_proposal\injuries\processed\injuries_cleaned_v1_4.csv")
```

```python
df['inj_norm_v4'] = (
    df['inj_norm_v3']
)

df.head(5)
```

```python
# TJ Related terms

tj_term_mapping = {
    'Elbow surgery (UCL)': 'Tommy John surgery',
    'Tommy John surgery/ flexor tendon surgery': 'Tommy John surgery',
    'Elbow surgery (UCL/flexor tendon repair)': 'Tommy John surgery',
    'Elbow surgery (UCL hybrid reconstruction)': 'Tommy John surgery'
}

tj_revision_mapping = {
    'Tommy John surgery (revision)': 'Tommy John surgery revision',
    'Tommy John revision surgery': 'Tommy John surgery revision',
    'Elbow surgery (UCL revision)': 'Tommy John surgery revision',
    'Elbow surgery (UCL revision repair)': 'Tommy John surgery revision',
    'Flexor tendon/Tommy John revision surgery': 'Tommy John surgery revision'
}

tj_internal_brace_mapping = {
    'Elbow surgery (UCL internal brace)': 'Tommy John internal brace surgery',
    'Elbow surgery (internal brace procedure)': 'Tommy John internal brace surgery',
    'Tommy John surgery (internal brace)': 'Tommy John internal brace surgery',
    'Elbow surgery (internal brace)': 'Tommy John internal brace surgery'
}

tj_internal_brace_revision_mapping = {
    'UCL revision surgery (internal brace)': 'Tommy John internal brace revision surgery'
}

tj_other_mapping = {
    'Tommy John surgery (non-throwing)': 'Tommy John surgery (non-throwing arm)'
}

tj_mapping = (
    tj_term_mapping |
    tj_revision_mapping |
    tj_internal_brace_mapping |
    tj_internal_brace_revision_mapping |
    tj_other_mapping
)

df["inj_norm_v4"] = (
    df["inj_norm_v4"]
        .astype("string")
        .str.strip()
        .replace(tj_mapping)
)

# 337
```

```python
# Update, check number of terms TEMPLATE
unique_injuries = df['inj_norm_v4'].unique()

print(f"Number of unique injury descriptions: {len(unique_injuries)}")

# Number of unique injury descriptions: 337
```
