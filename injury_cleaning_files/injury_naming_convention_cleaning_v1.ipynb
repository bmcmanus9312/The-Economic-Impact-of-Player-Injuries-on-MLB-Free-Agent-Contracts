```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\brend\Desktop\SABR_research_proposal\injuries\processed\injuries_2020_2025_clean.csv")

print(df.head(1))
```

```python
unique_injuries = df['inj_desc'].unique()

print(f"Number of unique injury descriptions: {len(unique_injuries)}")

# Number of unique injury descriptions: 762
```

```python
# Renaming Conventions

# manual naming changes for typos
df['inj_norm'] = (
    df['inj_desc'] # ONLY USE inj_desc for 1st assignment, then use NEWLY CREATED inj_norm for further changes
        .str.replace('sugery', 'surgery', regex=False)
        .str.replace('tendonitis', 'tendinitis', regex=False)
)

unique_injuries_v1 = df['inj_norm'].unique()

print(f"Number of unique injury descriptions: {len(unique_injuries_v1)}")

# Number of unique injury descriptions: 754
```

```python
# list of Infection related terms
filtered_entries = df['inj_norm'][df['inj_norm'].str.contains('infection', case=False, na=False)]
unique_infections = filtered_entries.unique() #
unique_infections_list = unique_infections.tolist()
print(unique_infections_list)
print(f"Number of unique infection-related injuries: {len(unique_infections)}")
```

```python

# 12 variations of 'infection'
convert_to_infection = [
    'Eye infection', 'Finger infection',
    'Leg infection', 'Viral infection',
    'Inner ear infection', 'Abdomen infection',
    'Calf infection', 'Knee infection',
    'Bacterial infection', 'Foot infection',
    'Kidney infection', 'Shoulder infection'
]

df['inj_norm'] = (
    df['inj_norm']
        .replace(convert_to_infection, 'Infection')
    )

# confirms one 'Infection' naming convention now
unique_injuries_v2 = df['inj_norm'][df['inj_norm'].str.contains('infection', case=False, na=False)].unique()
unique_infections_list_v2 = unique_injuries_v2.tolist()

print(unique_infections_list_v2)
print(f"Number of unique injury descriptions: {len(unique_injuries_v2)}")

```

```python

unique_injuries_v2 = df['inj_norm'].unique()
print(f"Number of unique injury descriptions: {len(unique_injuries_v2)}")

# Number of unique injury descriptions: 743
```

```python
# Illness related naming conventions
# list of Infection related terms
filtered_entries = df['inj_norm'][df['inj_norm'].str.contains('illness', case=False, na=False)]
unique_illnesses = filtered_entries.unique() #
unique_illnesses_list = unique_illnesses.tolist()
print(unique_illnesses_list)
print(f"Number of unique illness-related injuries: {len(unique_illnesses)}")
```

```python
# 4 variations of 'illness'
convert_to_illness = [
    'Illness', 'Non-COVID illness',
    'Non-viral illness', 'Viral illness'
]

df['inj_norm'] = (
    df['inj_norm']
        .replace(convert_to_illness, 'Illness')
    )

# confirms one 'Illness' naming convention now
unique_illnesses = df['inj_norm'][df['inj_norm'].str.contains('illness', case=False, na=False)].unique()
unique_illnesses_list = unique_illnesses.tolist()

print(unique_illnesses_list)
print(f"Number of unique injury descriptions: {len(unique_illnesses)}")
```

```python

unique_injuries_v3 = df['inj_norm'].unique()
print(f"Number of unique injury descriptions: {len(unique_injuries_v3)}")

# Number of unique injury descriptions: 740
```

```python
# COVID related naming conventions
# MUST BE COMPELTED AFTER NON-COVID ILLNESS RENAMING
filtered_entries = df['inj_norm'][df['inj_norm'].str.contains('COVID', case=False, na=False)]
unique_illnesses = filtered_entries.unique() #
unique_illnesses_list = unique_illnesses.tolist()
print(unique_illnesses_list)
print(f"Number of unique illness-related injuries: {len(unique_illnesses)}")
```

```python
# 11 variations of 'COVID-19'

convert_to_covid = [
    'COVID-19', 'COVID-19 protocol',
    'COVID-19 (protocol)', 'COVID-19 (cold symptoms)',
    'COVID-19 (vaccination)', 'COVID-19 (symptoms)',
    'COVID-19 (undisclosed)', 'COVID-19 IL (protocol)',
    'COVID-19 IL (undisclosed)', 'COVID-19 IL',
    'COVID-19 symptoms'
]

df['inj_norm'] = (
    df['inj_norm']
        .replace(convert_to_covid, 'COVID-19_related')
    )

# confirms one 'COVID-19' naming convention now
unique_covid = df['inj_norm'][df['inj_norm'].str.contains('COVID-19', case=False, na=False)].unique()
unique_covid_list = unique_covid.tolist()

print(unique_covid_list)
print(f"Number of unique injury descriptions: {len(unique_covid)}")
```
