```python
import pandas as pd

df = pd.read_csv(r"C:\Users\brend\Desktop\SABR_research_proposal\injuries\processed\injuries_cleaned_v1_3.csv")
```

```python
df['inj_norm_v3'] = (
    df['inj_norm_v2']
)

df.head(5)
```

```python
# Surgeries Mapping

# 'Hip surgery' canon
hip_surgery_mapping = {
    'Arthroscopic hip surgery': 'Hip surgery',
    'Hip surgery (torn labrum)': 'Hip surgery',
    'Hip surgery (surfacing procedure)': 'Hip surgery',
    'Hip surgery (labrum repair)': 'Hip surgery'
}

hand_surgery_mapping = {
    'Hamate surgery': 'Hand surgery',
    'Hamate surgery (right hand)': 'Hand surgery',
    'Hamate surgery (left hand)': 'Hand surgery'
}


['Core surgery; Microdiscectomy'], ['Abdominal surgery']

['Facial surgery'], ['Face surgery (OZMC fracture)'], ['Nasal fracture surgery', 'Nasal and facial fractures']


# eye
# ['Eye central serous chorioretinopathy', 'Laser eye surgery', 'Dry eye syndrome']

df["inj_norm_v3"] = (
    df["inj_norm_v3"]
        .astype("string")
        .str.strip()
        .replace(hip_surgery_mapping)
        .replace(hand_surgery_mapping)
)
```

```python
# Stress reaction injury mapping

# format 'Lower back stress reaction'
stress_reaction_mapping = {
    'Stress reaction in back': 'Lower back stress reaction',
    'Stress reaction to tibia': 'stress reaction',
    'Stress reaction (shin)': 'stress reaction',
    'Stress reaction -- shin': 'stress reaction',
    'Stress reaction -- shoulder': 'stress reaction',
    'Stress reaction -- arm': 'stress reaction',
    'Stress reaction in leg (fibula)': 'Leg stress reaction',
    'Stress reaction (toe)': 'Toe stress reaction',
    'Stress reaction in foot': 'Foot stress reaction',
    'Stress reaction in leg': 'Leg stress reaction',
    'Stress reaction in neck': 'Neck stress reaction',
    'Tibial stress reaction': 'Leg stress reaction',
    'Femoral stress reaction': 'Leg stress reaction',
}

nail_injury_mapping = {
    'Cracked fingernail': 'Finger discomfort',
    'Split fingernail': 'Finger discomfort',
    'Cracked nail (middle finger)': 'Finger discomfort',
    'Cracked fingernail/blister': 'Finger discomfort',
    'Fingernail avulsion (right middle)': 'Finger discomfort',
    'Fractured nail (right hand)': 'Finger discomfort'
}

df["inj_norm_v3"] = (
    df["inj_norm_v3"]
        .astype("string")
        .str.strip()
        .replace(stress_reaction_mapping)
        .replace(nail_injury_mapping)
)

# 416 unique
```

```python
# Lower back v2 injury mapping

lower_back_injury_mapping_update = {
    'Back spasms': 'Lower back spasms',
    'Strained back': 'Lower back strain',
    'Back tightness': 'Lower back discomfort',
    'Back stiffness': 'Lower back discomfort',
    'Strained back (trapezius)': 'Lower back strain',
    'Strained upper back': 'Lower back strain',
    'Back contusion': 'Lower back discomfort',
    'Back soreness': 'Lower back discomfort',
    'Ablation procedure (back)': 'Lower back surgery',
    'Strained upper back (rhomboid)': 'Lower back strain',
    'Lower back contusion': 'Lower back discomfort',
    'Stress fractures in back': 'Lower back stress fracture'
}

df["inj_norm_v3"] = (
    df["inj_norm_v3"]
        .astype("string")
        .str.strip()
        .replace(lower_back_injury_mapping_update)
)
# SKIPPED | 'Elbow/back soreness', 'Lower back disc herniation', 'Strained lower back/ Rib stress fracture', 'Herniated disc in back', 'Mid-back disc herniation'

# 404 unique
```

```python
# Tightness --> Discomfort

tightness_injury_mapping = {
    'Arm tightness': 'Arm discomfort',
    'Shoulder tightness': 'Shoulder discomfort',
    'Triceps tightness': 'Arm discomfort',
    'Lat tightness': 'Lower back discomfort',
    'Biceps tightness': 'Arm discomfort',
    'Hip tightness': 'Hip discomfort',
    'Thoracic spine tightness': 'Lower back discomfort',
    'Lower leg tightness': 'Leg discomfort'
}

soreness_injury_mapping = {
    'Glute soreness': 'Hip discomfort',
    'Thumb soreness': 'Finger discomfort',
    'Triceps soreness': 'Arm discomfort',
    'Shoulder soreness (frayed labrum)': 'Shoulder discomfort',
    'Biceps soreness': 'Arm discomfort',
    'Hand soreness': 'Hand discomfort'
}

bone_term_mapping = {
    'Elbow bone spur': 'Elbow discomfort',
    'Bone bruise on knee': 'Knee discomfort',
    'Bone bruise (finger)': 'Finger discomfort',
    'Bone bruise on hand (left)': 'Hand discomfort',
    'Bone bruise -- hand (right)': 'Hand discomfort',
    'Bone chip in elbow': 'Elbow discomfort',
    'Bone spur in elbow': 'Elbow discomfort',
    'Wrist bone bruise (left)': 'Wrist discomfort',
    'Knee bone bruise': 'Knee discomfort',
    'Bone bruise (wrist)': 'Wrist discomfort'
}

df["inj_norm_v3"] = (
    df["inj_norm_v3"]
        .astype("string")
        .str.strip()
        .replace(tightness_injury_mapping)
        .replace(soreness_injury_mapping)
        .replace(bone_term_mapping)
)

# 380 unique
```

```python
# Strain & Sprain terms v2

strain_injury_mapping = {
    'Strained hamstring': 'Strained leg',
    'Strained adductor': 'Strained hip',
    'Strained quad': 'Strained leg',
    'Strained lat': 'Strained lower back',
    'Lower back strain': 'Strained lower back',
    'Strained ribcage': 'Strained rib',
    'Shoulder strain': 'Strained shoulder',
    'Strained groin': 'Strained hip',
    'Strained flexor': 'Strained hip',
    'Strained biceps': 'Strained arm',
    'Strained intercostal': 'Strained rib',
    'Strained hamstring/quad': 'Strained leg',
    'Strained mass flexor': 'Strained forearm',
    'Strained teres major': 'Strained lower back',
    'Strained triceps': 'Strained arm',
    'Strained abdomen': 'Strained abdominal',
    'Strained cervical spine': 'Strained lower back',
    'Strained rotator cuff': 'Strained shoulder',
    'Strained right groin': 'Strained hip',
    'Strained glute': 'Strained hip',
    'Strained hip flexor': 'Strained hip ',
    'Strained thumb': 'Strained finger',
    'Strained shouler': 'Strained shoulder',
    'Strained gluteal': 'Strained hip'
}

sprain_injury_mapping = {
    'Sprained thumb': 'Sprained finger',
    'Sprained left thumb': 'Sprained finger',
    'Sprained ribcage': 'Sprained rib',
     'Sprained thumb (right)': 'Sprained finger',
     'Sprained thumb (left)': 'Sprained finger',
     'Sprained hand (left)': 'Sprained hand',
     'Sprained SC joint': 'Sprained shoulder',
     'Costochondral sprain': 'Sprained rib'
}

df["inj_norm_v3"] = (
    df["inj_norm_v3"]
        .astype("string")
        .str.strip()
        .replace(strain_injury_mapping)
        .replace(sprain_injury_mapping)

)

# SKIPPED | 'Strained quad/knee contusion', 'Strained calf, 'Strained pectoral','Strained oblique / Strained quad', 'Elbow soreness / Strained hamstring',
# SKIPPED | 'Strained lat/Torn shoulder capsule', 'Strained flexor tendon', 'Strained lower leg','Strained hand', 'Strained lower back/ Rib stress fracture'

# 354 unique
```

```python
# Torn & Rupture words

torn_injury_mapping = {
    'Torn ACL': 'Torn knee',
    'Torn hip flexor': 'Torn hip',
    'Torn lat': 'Torn lower back',
    'Torn finger tendon (right middle)': 'Torn finger',
    'Torn thumb ligament': 'Torn finger',
    'Torn toe ligament': 'Torn toe',
    'Ruptured finger pulley': 'Torn finger',
    'Ruptured thumb ligament': 'Torn finger',
    'Ruptured quad': 'Torn leg',
    'Torn hip labrum': 'Torn hip',
    'Thumb surgery (torn ligament)': 'Finger surgery'
}

df["inj_norm_v3"] = (
    df["inj_norm_v3"]
        .astype("string")
        .str.strip()
        .replace(torn_injury_mapping)
)

# 348 unique
```
