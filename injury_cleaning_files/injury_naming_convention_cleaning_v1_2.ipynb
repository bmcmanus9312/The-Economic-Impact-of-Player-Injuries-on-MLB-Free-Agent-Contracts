```python
### 2. term checker | change 'XXX' to desired term
filtered_entries = df['inj_norm_v2'][df['inj_norm_v2'].str.contains('disloc', case=False, na=False)]
unique_infections = filtered_entries.unique() #
unique_infections_list = unique_infections.tolist()
print(unique_infections_list)
print(f"Number of unique infection-related injuries: {len(unique_infections)}")
```

```python
# Surgery Name Corrections

neck_surgery_map = {
    'Neck surgery (disc replacement)': 'Neck surgery'
}

# all back classified as lower back for now, can be recleaned later if necesssary
lower_back_surgery_map = {
'Microdiscectomy (back surgery)': 'Lower back surgery',
'Back surgery (discectomy)': 'Lower back surgery',
'Back surgery': 'Lower back surgery',
'Back surgery (herniated disc)': 'Lower back surgery'
}

shoulder_surgery_map = {
    'Shoulder surgery (glenohumeral ligaments and capsu':'Shoulder surgery'
}

elbow_surgery_map = {
'Arthroscopic elbow surgery': 'Elbow surgery',
'Elbow surgery (bone chips)': 'Elbow surgery',
'Elbow surgery (stress fracture)': 'Elbow surgery',
'Arthoscopic elbow surgery': 'Elbow surgery',
'Elbow surgery (ulnar nerve transposition)': 'Elbow surgery'
}

forearm_surgery_map = {
    'Arm surgery (nerve-related)':'Forearm surgery'
}

hand_surgery_map = {
'Hand surgery (torn ligament)': 'Hand surgery',
'Hand surgery (hamate)': 'Hand surgery',
'Hand surgery (fractured hamate)': 'Hand surgery',
'Hand surgery (left hamate)': 'Hand surgery',
'Hand surgery (fractured metacarpal)': 'Hand surgery',
'Hand surgery (fractured hamate': 'Hand surgery',
'Hand surgery (non-throwing)': 'Hand surgery',
'Hand surgery (left)': 'Hand surgery'
}

finger_surgery_map = {
    'Finger surgery (fractured right ring)': 'Finger surgery',
    'Finger surgery (right index)': 'Finger surgery',
    'Finger surgery (ligament repair)': 'Finger surgery',
    'Finger surgery (right middle)': 'Finger surgery',
    'Finger surgery (left ring)': 'Finger surgery'
}


knee_surgery_map = {
'Knee Surgery': 'Knee surgery'
}

foot_surgery_map = {
'Foot Surgery': 'Foot surgery'
}

surgery_map = (
    neck_surgery_map |
    lower_back_surgery_map |
    shoulder_surgery_map |
    elbow_surgery_map |
    forearm_surgery_map |
    hand_surgery_map |
    finger_surgery_map |
    knee_surgery_map |
    foot_surgery_map
)

df["inj_norm_v2"] = (
    df["inj_norm_v2"]
        .astype("string")
        .str.strip()
        .replace(surgery_map)
)

# Update, check number of terms
unique_injuries = df['inj_norm_v2'].unique()

print(f"Number of unique injury descriptions: {len(unique_injuries)}")

# Number of unique injury descriptions: 531
```

```python
# Digit Name Corrections

# 'Fractured finger' canon
fractured_finger_map = {
    'Fractured finger (left pinky)': 'Fractured finger',
    'Fractured finger (right pinky)': 'Fractured finger',
    'Fractured pinky (left)': 'Fractured finger',
    'Fractured finger (right index)': 'Fractured finger',
    'Fractured finger (left index)': 'Fractured finger',
    'Fractured fingers (left ring, middle)': 'Fractured finger',
    'Fractured finger (right middle)': 'Fractured finger',
    'Fractured finger (left middle)': 'Fractured finger',
    'Fractured finger (left ring)': 'Fractured finger',
    'Fractured finger (right ring)': 'Fractured finger'
}

finger_surgery_map = {
    'Finger surgery (right index)': 'Finger surgery',
    'Finger surgery (right middle)': 'Finger surgery',
    'Finger surgery (fractured right ring)': 'Finger surgery',
    'Finger surgery (left ring)': 'Finger surgery',
}

finger_blister_map = {
    'Blister (right index finger)': 'Finger blister',
    'Blister on finger (left index)': 'Finger blister',
    'Blister -- finger (right index)': 'Finger blister',
    'Blister (index finger)': 'Finger blister',
    'Blister on finger (right index)': 'Finger blister',
    'Blister on finger (right middle)': 'Finger blister',
    'Blister (right middle finger)': 'Finger blister',
    'Blister (middle finger)': 'Finger blister',
    'Blister on finger': 'Finger blister',
    'Blister on finger (left)': 'Finger blister'
}

finger_symptom_map = {
    'Finger inflammation (left index)': 'Finger discomfort',
    'Finger inflammation (right index)': 'Finger discomfort',
    'Finger contusion (right middle)': 'Finger discomfort',
    'Finger inflammation (right middle)': 'Finger discomfort',
    'Finger discomfort (right middle)': 'Finger discomfort',
    'Finger contusion (middle right)': 'Finger discomfort',
    'Finger inflammation (right ring)': 'Finger discomfort'
}

finger_map = (
    fractured_finger_map |
    finger_surgery_map |
    finger_blister_map |
    finger_symptom_map
)

df["inj_norm_v2"] = (
    df["inj_norm_v2"]
        .astype("string")
        .str.strip()
        .replace(finger_map)
)

# 'Blister' is no location is given
blister_map = {
    'Blister on hand': 'Hand blister',
    'Blister on right hand': 'Hand blister',
    'Blisters': 'Blister',
    'Blister on toe': 'Toe blister',
    'Blister on foot': 'Foot blister',
    'Blisters (right hand)': 'Hand blister',
    'Blisters (left hand)': 'Hand blister'
}

df["inj_norm_v2"] = (
    df["inj_norm_v2"]
        .astype("string")
        .str.strip()
        .replace(blister_map)
)

# 500 here
```
